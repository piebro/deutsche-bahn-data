```python
import duckdb
import util

util.init()
```

```python
df = duckdb.sql("""
SELECT
    strftime(time, '%Y-%m') as months,
    AVG(delay_in_min) as "Durchschnittliche Versp채tung (min)",
    COUNT(*) as "Anzahl Stopps",
    SUM(CASE WHEN is_canceled THEN 1 ELSE 0 END) as "Ausgefallene Stopps"
FROM '../monthly_processed_data/*.parquet'
GROUP BY months
ORDER BY months
""").df()

util.show_figure(
    [
        util.FigureConfig(
            title="Monatliche Durchschnittsversp채tung",
            label="Durchschnittsversp채tung",
            x_col="months",
            y_col="Durchschnittliche Versp채tung (min)",
            query_or_df=df,
        )
    ]
)
```
