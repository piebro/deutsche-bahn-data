```python
import duckdb
import util

util.init()
```

```python
df = duckdb.sql("""
SELECT
    strftime(time, '%Y-%m') as months,
    AVG(delay_in_min) as "Average Delay (min)",
    COUNT(*) as "Total Stops",
    SUM(CASE WHEN is_canceled THEN 1 ELSE 0 END) as "Canceled Stops"
FROM '../monthly_processed_data/*.parquet'
GROUP BY months
ORDER BY months
""").df()

util.show_figure(
    [
        util.FigureConfig(
            title="Monthly Average Delay",
            label="Average Delay",
            x_col="months",
            y_col="Average Delay (min)",
            query_or_df=df,
        )
    ]
)
```
